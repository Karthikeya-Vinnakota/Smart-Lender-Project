```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv(r"C:\Users\shaik\Pictures\loan_prediction.csv")

print(data.head())
print(data.info())

```

```python
print(data['Gender'].unique())
print(data['Married'].unique())
print(data['Education'].unique())
print(data['Self_Employed'].unique())
print(data['Property_Area'].unique())
```

```python
plt.figure(figsize=(20,5))

# Gender vs Married
plt.subplot(131)
sns.countplot(x='Married', hue='Gender', data=data)

# Education vs Self Employed
plt.subplot(132)
sns.countplot(x='Self_Employed', hue='Education', data=data)

# Property Area vs Loan Amount Term
plt.subplot(133)
sns.countplot(x='Property_Area',
              hue='Loan_Amount_Term',
              data=data)

plt.tight_layout()
plt.show()
```
